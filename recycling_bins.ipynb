```python
import folium
from folium import plugins
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("recycle_bins.csv")
data.head()

data=data.dropna(subset=['Longitude'])

data=data.dropna(subset=['Latitude'])
```

```python
m = folium.Map([40.7128, -74.0060], tiles="Stamen Terrain", zoom_start=13)

for index, row in data.iterrows():
    folium.Marker([row['Latitude'], row['Longitude']], popup=folium.Popup(row['Address'], parse_html=True)).add_to(m)
m
```
